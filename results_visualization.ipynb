```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("results.csv")
```

```python
results.columns = results.columns.str.replace(' ', '')
results.head()
```

```python
NUM_EPOCHS = 300
fig, ax = plt.subplots()
print(results.columns)
results.plot(x="epoch", y="metrics/precision", kind="line", ax=ax)
results.plot(x="epoch", y="metrics/recall", kind="line", ax=ax)
results.plot(x="epoch", y="metrics/mAP_0.5", kind="line", ax=ax)
results.plot(x="epoch", y="metrics/mAP_0.5:0.95", kind="line", ax=ax)
ax.set_title("Training performance")
```

```python
fig, ax = plt.subplots()
results.plot(x="epoch", y="train/box_loss", kind="line", ax=ax)
results.plot(x="epoch", y="train/obj_loss", kind="line", ax=ax)
results.plot(x="epoch", y="val/box_loss", kind="line", ax=ax)
results.plot(x="epoch", y="val/obj_loss", kind="line", ax=ax)
ax.set_title("Losses")
```
